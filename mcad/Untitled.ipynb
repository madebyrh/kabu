```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('mcad.csv', encoding='sjis')
df
```

## MCADの9日移動平均を求める

```python
df_e = pd.DataFrame()
df_e['SC'] = df['SC']
df_e['名称'] = df['名称']
df_e['prev_signal'] = df.loc[:, df.columns[-10]:df.columns[-2]].mean(axis=1)
df_e['latest_signal'] = df.loc[:, df.columns[-9]:df.columns[-1]].mean(axis=1)
df_e
```

```python
df_e['diff'] = df_e['prev_signal'] - df_e['latest_signal']
df_e['flg'] = df_e['diff'].apply(lambda x: 1 if x > 0 else 0)
df_e
```
